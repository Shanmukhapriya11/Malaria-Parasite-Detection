```python
from PIL import Image
import os
import matplotlib.pyplot as plt

# Define the directory containing your images
image_dir = "C:/Users/Psps0/Downloads/archive/Malaria Cells/training_set"

# List all image files in the directory
image_files = [f for f in os.listdir(image_dir) if os.path.isfile(os.path.join(image_dir, f))]

# Data cleaning
cleaned_images = []
for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)
    try:
        # Attempt to open the image file
        with Image.open(image_path) as img:
            # Example: Resize images to 256x256
            img_resized = img.resize((256, 256))
            # Add resized image to the cleaned images list
            cleaned_images.append(img_resized)
    except Exception as e:
        # If unable to open the image, print an error message
        print(f"Error processing image {image_path}: {e}")

# Save cleaned images (optional)
output_dir = "cleaned_images/"
os.makedirs(output_dir, exist_ok=True)
for i, img in enumerate(cleaned_images):
    img.save(os.path.join(output_dir, f"image_{i}.jpg"))

print("Data cleaning completed.")
```

```python
import os
import cv2
import matplotlib.pyplot as plt

# Function to preprocess an individual image
def preprocess_image(image_path):
    # Read the image
    image = cv2.imread(image_path)
    
    # Convert to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    return image, grayscale_image

# Function to preprocess an entire dataset and display images in plots
def preprocess_and_display(dataset_dir, num_images=3):
    # List all image files in the dataset directory
    image_files = [f for f in os.listdir(dataset_dir) if f.endswith('.jpg') or f.endswith('.png')]
    
    # Preprocess and display a subset of images
    for image_file in image_files[:num_images]:
        image_path = os.path.join(dataset_dir, image_file)
        original_image, grayscale_image = preprocess_image(image_path)
        
        # Plot original and grayscale images
        plt.figure(figsize=(8, 4))
        
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
        plt.title('Original Image')
        plt.axis('off')
        
        plt.subplot(1, 2, 2)
        plt.imshow(grayscale_image, cmap='gray')
        plt.title('Grayscale Image')
        plt.axis('off')
        
        plt.show()

# Example usage:
dataset_dir = 'C:/Users/Psps0/Downloads/archive/Malaria Cells/training_set/Uninfected'
preprocess_and_display(dataset_dir, num_images=5)
```

```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

img_width, img_height = 140, 140
batch_size = 34

train_dir = "C:/Users/Psps0/archive/Malaria Cells/training_set"
validation_dir = "C:/Users/Psps0/archive/Malaria Cells/single_prediction"
test_dir = "C:/Users/Psps0/archive/Malaria Cells/testing_set"
```

```python
# Data preprocessing and augmentation for training dataset
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary'
)

# Data preprocessing for validation dataset
validation_datagen = ImageDataGenerator(rescale=1./255)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary'
)

# Data preprocessing for testing dataset
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)
```

```python
# Define the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size
)
```

```python
# Evaluate the model on the testing dataset
test_loss, test_accuracy = model.evaluate(test_generator, verbose=2)
print(f"Test accuracy: {test_accuracy}")

```

```python
# Make prediction on a random image from the testing dataset
random_image_batch, random_label_batch = test_generator[0]
random_index = random.randint(0, batch_size-1)
random_image = random_image_batch[random_index]
random_label = random_label_batch[random_index]
random_image_expanded = np.expand_dims(random_image, axis=0)
```

```python
# Make prediction
prediction = model.predict(random_image_expanded)
predicted_class = "Infected" if prediction[0][0] > 0.5 else "Uninfected"

```
